## 2. Пропуски и очистка
2.1) Стратегия dropna()

Пропусков мало, поэтому удалим их

```python
df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True)
```

2.2) Стратегия fillna()

Заполнение константой (так как их нельзя угадать)

```python
df['director'] = df["director"].fillna("Unknown")
df['cast'] = df["cast"].fillna("No info")
```

Заполнение модой (так как Netflix чаще всего выпускает фильмы в одной стране)

```python
country_mode = df['country'].mode()[0]
df['country'] = df["country"].fillna(country_mode)
```

3.3) Дисперсия, ассиметрия, эксцесс

Дисперсия - это то, как разбросаны данные относительно среднего значения. 

В моем случае показатель 77 значит, что данные однородны.

```python
df['release_year'].var()
```

Ассиметрия - это то, как симметрично относительно среднего распределены данные. 

В моем случае отрицательное значение показывает, что большинство значений выше среднего.

```python
df['release_year'].skew()
```

Эксцесс - это то, насколько острый или плоский пик у распределения.

В моем случае такой показатель означает, что большинство данных находятся в очень узком диапазон значение возле среднего.

```python
df['release_year'].kurtosis()
```

## 4. Энкодинг и инжиниринг

```python
print("Датасет до манипуляция")
df.head(15)
```

4.1) Создание новых признаков

Разобьем дату добавления

```python
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df.head(15)
```

4.2) One-Hot encoding

```python
df_encoded = pd.get_dummies(df, columns=['type'], prefix='type')
df_encoded.head(10)
```

4.3) Feature Hashing

```python
from sklearn.feature_extraction import FeatureHasher

hasher = FeatureHasher(n_features=5, input_type='string')
hashed_features = hasher.transform(df['country'].apply(lambda x: [x])).toarray()
hash_columns = [f'country_hash_{i}' for i in range(5)]
df_hashed = pd.DataFrame(hashed_features, columns=hash_columns, index=df.index)
df_hashed.head(10)
```
